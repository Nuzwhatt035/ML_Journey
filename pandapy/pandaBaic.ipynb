```python
import pandas as pd

students = {
    "Name": ["Nuzhat", "Iffat","Iffu", "NUz", "Jeelani"],
    "Branch": [ "cse", "mechanical", "ai", "civil", "electronics"],
    "Marks": [ 99, 100, 20, 30 ,40]
}

df = pd.DataFrame(students)
print(df.groupby("Branch")["Marks"].mean())
    
    
```

```python
import pandas as pd

employees = {
    "Name": ["Nuzhat", "Iffat","Iffu", "NUz", "Jeelani"],
    "Department": ["Sales", "HR", "Sales", "IT", "HR"],
    "Salary": [50000, 60000, 55000, 70000, 65000]
}

df = pd.DataFrame(employees)

print(df.groupby("Department")["Salary"].sum())
```

```python
import pandas as pd

movies = {
    "Movie": ["PK", "Holiday", "Avengers", "Race", "Golmal"],
    "Genre": ["comedy", "rom-com", "action", "drama", "retro"],
    "Rating": [ 5, 4, 3, 2, 1]
}

df = pd.DataFrame(movies)

print(df.groupby("Genre")["Rating"].max())
```

```python
#SLIGHTLY CHALLENGING

import pandas as pd

fitness = {
    "Age": [19, 20, 21, 22, 23, 24],
    "WorkoutType": [ "walking", "cycling", "running", "swimming", "squats", "push-ups"],
    "Calories": [400, 500, 600, 700, 800, 900]

}

df = pd.DataFrame(fitness)

print(df.groupby("WorkoutType")["Calories"].mean())
print(df.groupby("WorkoutType")["Calories"].max())



    
```
