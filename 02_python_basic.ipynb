```python
def grade_statistics(marks):
    highest_mark = marks[0]
    lowest_mark = marks[0]
    total_sum = 0

    for num in marks:
        if num > highest_mark:
           highest_mark = num

        if num < lowest_mark:
           lowest_mark = num

        total_sum += num

    average_mark = total_sum/ len(marks)

    return highest_mark,lowest_mark,average_mark

high,low,avg = grade_statistics( [65, 80, 72, 90, 55])

print(high,low,avg)
```

```python
def analyze_marks(marks):
    highest_mark = marks[0] #we use square brackets for indexing
    lowest_mark = marks[0]
    Total_Students = len(marks)
    total_Sum = 0

    for num in marks:
        if highest_mark < num:
            highest_mark = num

        if lowest_mark > num:
            lowest_mark = num

        total_Sum += num

    average_mark = total_Sum/Total_Students

    return highest_mark,lowest_mark,Total_Students, average_mark



Highest , Lowest , Number_of_Students , Average = analyze_marks((78, 92, 85, 67, 95))

print(Highest , Lowest , Number_of_Students , Average )
```

```python
def analyze_students(students):

    # Get the first student and marks
    first_name = list(students.keys())[0]
    highest_mark = students[first_name]
    top_student = first_name

    total_sum = 0

    for name, mark in students.items():

        if mark > highest_mark:
            highest_mark = mark
            top_student = name

        total_sum += mark

    average_mark = total_sum / len(students)

    return top_student, highest_mark, average_mark


TopStudent, HighestMark, AverageMark = analyze_students({
    "Ali": 85,
    "Sara": 91,
    "John": 78,
    "Nuzhat": 95
})

print("Top Student:", TopStudent)
print("Highest Mark:", HighestMark)
print("Average Mark:", AverageMark)
```

```python
class Rectangle:
    def __init__(self, length, breadth):
        # We store the dimensions as attributes of the object
        self.length = length
        self.breadth = breadth

    def area(self):
        # Now the method uses the stored attributes
        return self.length * self.breadth

# Setup: The object remembers its size
rect = Rectangle(10, 8)

# Calculation: No arguments needed here because the object knows its dimensions
print(f"Area: {rect.area()}")
```
