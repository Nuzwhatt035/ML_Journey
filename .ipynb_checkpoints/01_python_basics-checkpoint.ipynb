```python
#calculator eg
x = int (input("enter first number: "))
y = int (input("enter second number: " ))
option = input( "please enter your choice(+,-,*,/,%) : ")
print(option)
if option == "+":
    print( "sum:", x + y)
elif option == "-":
    print( "difference:", x - y)
elif option == "*":
    print("product:",  x * y)
elif option == "/":
    #to prevent crashing we will avoid zero
    if y == 0:
        print("oops!you used zero")
    else:
        print("quotient:",  x / y)
elif option == "%":
    print("remainder:",  x % y)
else:
 print("oops! invalid choice")
```

```python
#for-loop in list
fruits = ["Apple", "Banana", "Mango"]

for fruit in fruits:
    print(fruit)
```

```python
#Using len() with range()
fruits = ["Apple", "Banana", "Mango"]
for i in range(len(fruits)):
    print(i, fruits[i])
```

```python
#while loop
count = 1
while count <= 5:
    print(count)
    count += 1
```

```python
#break
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
#continue
for i in range(6):
    if i == 3:
        continue
    print(i)
```

```python
#nested loop
for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
#enumerate()
fruits = ["Apple", "Banana", "Mango"]

for index, fruit in enumerate(fruits):
    print(index, fruit)
```
